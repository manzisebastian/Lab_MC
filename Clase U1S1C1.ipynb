#### Operaciones con vectores

Al igual que en Álgebra, los vectores se pueden sumar, restar, multiplicar (entre sí y por un escalar) y dividir, aunque con excepciones.

```python
# Creamos cuatro vectores cualesquiera

# 2 vectores-fila
t = np.array([1, 2, 3])
u = np.array([4, 5, 6])

# 2 vectores-columna
v = np.array([[1], [2], [3]]) # shape (3,1)
w = np.array([[4], [5], [6]]) # shape (3,1)
```

Los vectores-fila de igual dimensión pueden sumar, restar, multiplicar y dividir *elemento a elemento*.

```python
# Para vectores-fila
print("t + u = ", t + u)
print("t - u = ", t - u)
print("t * u = ", t * u)
print("t / u = ", t / u)
```

```python
# Para vectores-columna
print("v + w = ", v + w)
print("v - w = ", v - w)
print("v * w = ", v * w)
print("v / w = ", v / w)
```

**OJO**

En Álgebra, los vectores-fila y vectores-columna **no** se podían operar conjuntamente.

En cambio, NumPy **sí** permite esto.

```python
# Para vectores-fila y vectores-columna
print("t + v = ", t + v)
print("\n\nt - v = ", t - v)
print("\n\nt * v = ", t * v)
print("\n\nt / v = ", t / v)
```

Vemos que el resultado es...una matriz de *3x3*, en todos los casos.

Después vamos a explorar las matrices. Primero, entendamos por qué sucede esto.

Cuando realizamos operaciones entre arrays en NumPy, sus dimensiones (`shapes`) se comparan **desde la última dimensión hacia la izquierda** aplicando la regla de *broadcasting*.

```python
print(t.shape)
print(v.shape)
```

Un vector con `shape (3,)` en realidad se interpreta como `(1,3)` (un array de 1 fila y 3 columnas), mientras que un vector con `shape (3,1)` es una columna de 3 filas y 1 columna. Al compararlos:

- `(1,3)` contra `(3,1)`
- La última dimensión: `3` vs `1` → compatibles porque una de ellas es 1 → resultado `3`.
- La penúltima dimensión: `1` vs `3` → también compatibles → resultado `3`.

Así, NumPy “expande” el primero hacia abajo (replica la fila 3 veces) y el segundo hacia la derecha (replica la columna 3 veces). El resultado final es un array de `shape (3,3)`, donde cada entrada `(i,j)` combina el elemento `t[j]` con el elemento `v[i]`.

Lo mismo vale para el caso en el que se suma un vector-columna y un vector-fila (`u + w`).

### Otras operaciones con vectores

Al igual que en Álgebra, los vectores (fila o columna, indistintamente) pueden multiplicarse por un escalar.

```python
# Definimos un escalar, k
k = 1.5
```

```python
# Para un vector-fila
t_ = t * k
print(t_)
```

```python
# Para un vector-columna
v_ = v * k
print(v_)
```

También podemos obtener el producto-punto de dos vectores (fila, en este caso).

```python
# Para obtener el producto-punto, procedemos de cualquiera de las 2 siguientes maneras
dot_product_1 = u @ t
print(dot_product_1)
dot_product_2 = np.dot(u, t)
print(dot_product_2)
```

## 3) Visualización básica con Matplotlib

Documentación: https://matplotlib.org/stable/

Graficamos funciones lineales y cuadráticas.

```python
# Importamos la librería Matplotlib para graficar funciones
# y la renombramos como "plt" para que sea más fácil
import matplotlib.pyplot as plt
```

Vamos a definir 2 funciones (una lineal y otra cuadrática) para visualizarlas en un gráfico de Matplotlib.

Definimos los valores del dominio (sobre los que se *plotteará* la función).

Para ello, usamos `np.linspace()`:
- Documentación: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
- Sintaxis:
`
numpy.linspace(start, stop, num= ,...)`


```python
xs = np.linspace(-10, 10, 200)
```

```python
# Pendiente e intercepto
m, b = 2, 1  # pendiente e intercepto
```

```python
# Funciones
ys_lineal = m * xs + b
ys_cuad = xs**2 - 4*xs + 1
```

```python
# Creamos una figura
plt.figure(figsize=(7,4))

# Incluímos la función lineal
plt.plot(xs,
         ys_lineal,
         label='Recta: m*x + b')

# Incluímos la función cuadrática
plt.plot(xs,
         ys_cuad,
         label='Cuadrática: x^2 - 4x + 1')

# Marcamos los ejes
plt.axhline(0)
plt.axvline(0)

# Agregamos un título
plt.title('Funciones lineal y cuadrática')

# Agregamos etiquetas
plt.xlabel('x'); plt.ylabel('y')

# Formamos y mostramos la figure
plt.legend(); plt.grid(alpha=0.3)
plt.show()
```
